```python
import pandas as pd
import json
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.graph_objs as go
import pandas as pd
import csv
import plotly.plotly as py
from dash.dependencies import Input, Output
from datetime import datetime as dt
import datetime
```

```python
with open('events.json') as f:
    events = json.load(f) # events ist dann die Liste mit Dictionaries

for event in events:
    print(event)
```

```python
df = pd.DataFrame(events)
df
```

```python
#start = datetime.datetime(2018, 1, 1 )
#end = datetime.datetime.now

app = dash.Dash()

app.layout = html.Div(children=[
    html.Div(children='''
            Test Report
    '''),
    
    dcc.Dropdown(
        id='camera-dropdown',
        options=[
            {'label': 'webcam_1', 'value': 'webcam_1'},
            {'label': 'webcam_2', 'value': 'webcam_2'},
            {'label': 'webcam_3', 'value': 'webcam_3'}
        ],
        value='webcam_1'
    ),
    dcc.Dropdown(
        id='event-dropdown',
        options=[
            {'label': 'face', 'value': 'face'},
            {'label': 'frame', 'value': 'frame'},
            {'label': 'movement', 'value': 'movement'}
        ],
        value='frame'
    ),
    html.Div(children='''
        Here you can see the report test.
    '''),

    dcc.Graph(
        id='camera-graph',
        figure={
            'data': [
                    {'x':'df.timestamp', 'y':'df.Anzahl_ID', 'text':'df.event_type', 'type':'line', 'name' : 'IDK', 'mode':'markers'}
                
            ]
        }
    ),
])
@app.callback(
    Output('camera-graph', 'figure'),
    [Input('event-dropdown', 'value'),
     Input('camera-dropdown', 'value')])
def update_graph(selected_dropdown_value1, selected_dropdown_value2):
    dff = df[df.camera_id == selected_dropdown_value2]
    dfff= dff[dff.event_type == selected_dropdown_value1]
    return{
        'data':[go.Scatter(
            x=dfff['timestamp'],
            y=dfff['Anzahl_ID'],
            text=dfff['event_type'],
            name= 'IDK',
            type='line',
            mode='markers')
        ],
            'layout': {
                'title' : 'Reporting cameras'
            }
    }
    

if __name__ == '__main__':
    app.run_server()
```
